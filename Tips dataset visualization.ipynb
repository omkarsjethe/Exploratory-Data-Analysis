```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
sns.set(style="white")

# Generate a random univariate dataset
rs = np.random.RandomState(10)
df = rs.normal(size=100)

# Plot a simple histogram and kde
sns.histplot(df, kde=True, color="m")
```
